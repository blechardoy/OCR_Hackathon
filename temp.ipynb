```python
import os
import json

def extract_rf_texts(annotations_dir):
    """
    Itère sur tous les fichiers JSON dans le dossier annotations et affiche les valeurs de "text"
    pour les objets ayant le label "RF".
    
    Arguments :
    - annotations_dir (str) : Chemin du dossier contenant les fichiers JSON d'annotations.
    """
    # Vérification du dossier
    if not os.path.exists(annotations_dir):
        print(f"Le dossier {annotations_dir} n'existe pas.")
        return
    
    # Itération sur tous les fichiers JSON dans le dossier
    for filename in os.listdir(annotations_dir):
        if filename.endswith(".json"):
            file_path = os.path.join(annotations_dir, filename)
            
            # Lecture du fichier JSON
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            # Vérification de la clé "form" dans le JSON
            if "form" in data:
                for item in data["form"]:
                    # Vérification du label
                    if item.get("label") == "RF":
                        # Récupération et affichage de la valeur de "text"
                        text = item.get("text", "Aucun texte trouvé")
                        print(f"{text}")

# Chemin du dossier annotations
annotations_dir = "data/train/annotations"
extract_rf_texts(annotations_dir)
```
